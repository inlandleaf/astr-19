# Unit 2 Code Demo

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*(np.pi), 100) # x array
y = np.sin(5*x) # y array
```

```python
plt.plot(x, y) # plot x vs y
plt.ylim([-1.1, 1.1]) # limit y axis

# label axes
plt.xlabel('x')
plt.ylabel('sin(5x)')

plt.savefig('demo.png', facecolor = 'lightgray')
```
